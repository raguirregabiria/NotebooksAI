Obtenemos la pagina de la cartelera

```python
import urllib.request as ur
f = ur.urlopen("https://www.ecartelera.com/cines/56,0,1.html")
s = f.read().decode()
f.close()
```

Creamos la expresión regular que obtiene el titulo de la pelicula y el director

```python
regex='alt = "(.*?)\".*?Dir\..*?hostlink\'>(.*?)<'
```

Ejecutamos la expresión

```python
import re
matches = re.findall(regex,s, re.DOTALL)
print(matches)
```

Esta celda de abajo es una función que cambia los espacios por '+' (recordemos que en las URL el espacio se identifica con '+')

```python
def masPorEspacio(nombre):
    resultado=""
    for i in nombre:
        if i==" ":
            resultado+="+"
        else:
            resultado+=i
    return resultado

print(masPorEspacio("Hola que hay"))
```

La siguiente función sirve para quitar acentos y eñes, esto es importante porque para la función urlopen, hay que pasarle una cadena en ASCII. Hay librerías que hacen esto (como unidecode), pero para evitar la instalación preferimos definir esta sencilla función, modificada de una similar https://micro.recursospython.com/recursos/como-quitar-tildes-de-una-cadena.html

```python
def normalize(s):
    replacements = (
        ("á", "a"),
        ("é", "e"),
        ("í", "i"),
        ("ó", "o"),
        ("ú", "u"),
        ("ñ","n")
    )
    for a, b in replacements:
        s = s.replace(a, b).replace(a.upper(), b.upper())
    return s

print(normalize("¡malasaña!"))
print(normalize("¡HÓLÁ, MÚNDÓ!"))
```

En la celda inferior, recorremos todas la películas encontradas anteriormente, haciendo una búsqueda por cada una y obteniendo una pagina de la cual sacamos la nota de cada una. Usamos filaffinity porque IMDB presentaba problemas a la hora de buscar por título y director.

```python
notas=[]
for pelicula in matches:
    consulta="https://www.filmaffinity.com/es/advsearch.php?stext="+normalize(masPorEspacio(pelicula[0]))+"+"+normalize(masPorEspacio(pelicula[1]))+"&stype%5B%5D=title&stype%5B%5D=director&country"
    f1 = ur.urlopen(consulta)
    s1 = f1.read().decode()
    f1.close()
    regexNota="avgrat-box\">(.*?)<.*"
    notas.append(re.findall(regexNota,s1, re.DOTALL))
print(notas)
    
```

Esta expresión encuentra el trozo de texto que queremos reemplazar (que es la nota). Ojo esta es la nota de la propia eCartelera, que es distinta a la de filmaffinity.

```python
regexReplace='alt = \"(.*?)<\/span>'
```

```python
matchesReplace=re.findall(regexReplace, s, re.DOTALL)
print(matchesReplace)
   
```

Aquí llamamos a la función replace que busca los string que queremos sustituir y los remplaza por el mismo menos los tres últimos caracteres (que son los que contienen la nota de eCartelera). En estos introduciomos la nota encontrada en filmaffinity.

```python
for i in range(0,len(matchesReplace)):
    s=s.replace(matchesReplace[i],matchesReplace[i][:-3]+notas[i][0])
print(s)

```

Ahora que tenemos el html actualizado, los guardamos y cuando lo abramos, veremos la página con las notas de filmaffinity.

```python
archivo=open("NotasCine.html","w")
archivo.write(s)
archivo.close()
```

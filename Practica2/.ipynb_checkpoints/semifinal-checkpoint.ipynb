```python
import numpy as np

class Perceptron:
    def __init__(self, ninput, noutput):
        self.ninput = ninput
        self.noutput = noutput
        self.w = np.random.rand(ninput,noutput)-0.5
        self.b = np.random.rand(noutput)-0.5
        
    def forward (self, x):
        neta = x.dot(self.w)+self.b
        resultado = neta
        for i in range(len(neta)):
            
            if neta[i]>0:
                resultado[i]=1
            else:
                resultado[i]=0
        return resultado
        '''salida = 1/(1+np.exp(-neta))
        
        return salida'''

        
    def update (self, x, d, alpha):
        s = self.forward(x) 
        
        lista = [self.b[0]]
        for i in self.w:
            lista.append(i[0])
        pesos = np.array(lista)[np.newaxis].T
        
        x =  x[np.newaxis].T
        lista= [1.0]
        for i in x:
            lista.append(i[0])
        entradas = np.array(lista)[np.newaxis].T
        variacion = alpha * entradas.dot(d-s)

        pesos = pesos + variacion[np.newaxis].T

        self.b = np.array(pesos[0])
   
        self.w = np.array([pesos[1],pesos[2]])
    
   

        
        
    def RMS (self, X, D): # calcula el error RMS
        S = self.forward(X)
        return np.mean(np.sqrt(np.mean(np.square(S-D),axis=1)))
        
    def accuracy (self, X, D): # calcula el ratio de aciertos
        S = self.forward(X)
        errors = np.mean(np.abs(D-S))
        return 1.0 - errors
    
    def info (self, X, D): # traza de cómno va el entrenamiento
        print('     RMS: %6.5f' % self.RMS(X,D))
        print('Accuracy: %6.5f' % self.accuracy(X,D))
        
    def train (self, X, D, alpha, epochs, trace=0): # entrena usando update
        for e in range(1,epochs+1):
            for i in range(len(X)):
                self.update(X[i],D[i], alpha)
            if trace!=0 and e%trace == 0:
                print('\n   Epoch: %d' % e)
                self.info(X,D)
                

```

```python
class Multilayer:
    def __init__(self, ninput, nhidden, noutput):
        self.ninput = ninput # número de entradas
        self.nhidden = nhidden # número de neuronas en capa oculta
        self.noutput = noutput # número de salidas
        self.salida1 = []
        self.w1 = np.random.rand(ninput,nhidden)-0.5
        self.b1 = np.random.rand(nhidden)-0.5
        self.w2 = np.random.rand(nhidden,noutput)-0.5
        self.b2 = np.random.rand(noutput)-0.5
        self.testPercentage = 10
        
#         self.w1 = np.array([[-1.0,0.0],[0.0,1.0]])
#         self.b1 = np.array([1.0,2.0])
#         self.w2 = np.array([[0.5],[-0.5]])
#         self.b2 = np.array([1.0])
        
        self.lRMS = [] # contiene la lista de RMSs para pintarlos luego
        self.laccuracy = [] # contiene la lista de accuracy para pintar luego
        


    def sigm (self, neta): # función sigmoidal
        return 1.0 / (1.0 + np.exp(-neta))
    
    def forward (self, x): # propaga un vector x y devuelve la salida         

        neta = x.dot(self.w1)+self.b1
        salida = self.sigm(neta)
        self.salida1 = salida

        neta = salida.dot(self.w2)+self.b2
        salida = self.sigm(neta)

        return salida
    
    def update (self, x, d, alpha): # realiza una iteración de entrenamiento
        # a implementar

        s = self.forward(x)
        delta2 = (d-s).dot(s.dot(1 - s))
        
        delta1 =  delta2.dot(self.w2.reshape(self.noutput,self.nhidden)*(self.salida1*(1-self.salida1)))
        incw1 = alpha * x.reshape(self.ninput,1)*delta1
        self.b1 = self.b1 + delta1 * alpha
        
        incw2 = alpha*self.salida1.reshape(self.nhidden,1).dot(delta2.reshape(1,self.noutput))
   
        self.b2 = self.b2 + delta2 * alpha
        self.w1 += incw1
        self.w2 += incw2
#         print("Delta 2\n")
#         print(delta2)
#         print("Delta 1\n")
#         print(delta1)
#         print("W 1\n")
#         print(self.w1)
#         print("B 1\n")
#         print(self.b1)
#         print("W 2\n")
#         print(self.w2)
#         print("B 2\n")
#         print(self.b2)

    
    def RMS (self, X, D): # error RMS
        S = self.forward(X)
        return np.mean(np.sqrt(np.mean(np.square(S-D),axis=1)))
        
    def accuracy (self, X, D): # calcula ratio de4 aciertos
        S = self.forward(X)
        S = np.round(S)
        errors = np.mean(np.abs(D-S))
        return 1.0 - errors
    
    def info (self, X, D): # escribe traza
        self.lRMS.append(self.RMS(X,D))
        self.laccuracy.append(self.accuracy(X,D))
        print('     RMS: %6.5f' % self.lRMS[-1])
        print('Accuracy: %6.5f' % self.laccuracy[-1])
        
    def train (self, X, D, alpha, epochs, trace=0): # entrena usando update
        self.lRMS = [] # guarda lista de RMSs para pintarlos
        self.laccuracy = [] # guarda lista de accuracy
        for e in range(1,epochs+1):
            for i in range(len(X)):
                self.update(X[i],D[i], alpha)
            if trace!=0 and e%trace == 0:
                print('\n   Epoch: %d' % e)
                self.info(X,D)
    def trainTest (self, X, D, alpha, epochs, trace=0, percentageTest=10): # Define el porcentaje del set que se quiere como test
               
        self.lRMS = [] # guarda lista de RMSs para pintarlos
        self.laccuracy = [] # guarda lista de accuracy
        self.testPercentage=(int)(len(X)*0.01*percentageTest)

        for e in range(1,epochs+1):
            for i in range(len(X)-self.testPercentage):
                self.update(X[i],D[i], alpha)
            if trace!=0 and e%trace == 0:
                print('\n   Epoch: %d' % e)
                self.testError(X,D)
    def testError (self, X, D):        
        print("Sobre el conjunto de entrenamiento:")
        self.info(X[:len(X)-self.testPercentage],D[:len(D)-self.testPercentage])
        print("Sobre el conjunto de test:")
        self.info(X[-self.testPercentage:],D[-self.testPercentage:])
        
                
def one_hot (d): # codificación one_hot
    num_classes = len(set(d))
    rows = d.shape[0]
    labels = np.zeros((rows, num_classes), dtype='float32')
    labels[np.arange(rows),d.T] = 1
    return labels
```
